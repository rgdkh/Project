# Data Mining IT326 - Project 
# Group 4  (Home Loan Predictions)

# Problem:

We chose this dataset, in light of the increasing challenges associated with registering loans for housing. The company embarked on a mission to use a specific data set to predict which individuals meet the basic requirements for loans. This prediction is based on various attributes present in the data set and factors such as marital status, self-employment status, number of dependents, and more. The primary goal is to ensure that loans are provided to applicants who are highly likely to meet their financial obligations. This endeavor is critical because it aims to reduce the risk of loan defaults and other related problems, an issue of paramount importance in the current lending landscape.

Dataset:https://www.kaggle.com/datasets/gavincanacam/home-loan-predictions?datasetId=246760


# Dataset information:

Number of Objects: 615

Number of Attributes: 12

|Attribute|Description|Type|Possible Values|
|-----|-----|-----|----|
|Loan_ID| the client' ID| Nominal |From LP001002 toLP002990|
|Gender|the gender of the client| Binary | Female - Male|
|Married|if the client married or not| Binary | Yes - No|
|Dependents|if the client has children| Numeric| From 0  to +3 |
|Education|education statue of the client| Binary | Not Graduate - Graduate |
|Self_Employed|if the client has a job | Binary | Yes - No |
|ApplicantIncome|how much is the Income for the client| Numeric | From 150 to 81000 |
|CoapplicantIncome|how much Income does the co-applicant have| Numeric | From 0 to 41667 |
|LoanAmount|how much is the amount of the loan| Numeric | From 9 to 700 |
|Loan_Amount_Term|the amount of time the lender gives the client to repay the loan| Numeric | From 12 to 480 |
|Credit_History|the client history with loans| Binary | 0(No) - 1(Yes) |
|Property_Area|the client area| Nominal | Urban - Rural - Semiurban |
|Loan_Status|does the client can have a loan or not | Binary | Yes - No |
